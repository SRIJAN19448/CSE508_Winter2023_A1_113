# Q1

## Relevant Text Extraction

```python
os.makedirs('Q1_1_output', exist_ok=True)
base="CSE508_Winter2023_Dataset"
sample=os.listdir(base)[:5]
for filename in os.listdir(base):
    with open(base+"/"+filename, 'r+') as file1:
        contents = file1.read()
        soup=BeautifulSoup(contents)
        title=soup.find('title')
        text=soup.find('text')
        temp=title.text+" "+text.text
        # file1.seek(0)
        # file1.truncate()
        with open("Q1_1_output"+"/"+filename, 'w') as file2:
            file2.write(temp)
        if(filename in sample):
            print("\033[91mBefore~~~~~~~~~~~~~~~\033[0m\n")
            print(soup)
            print()
            print("\033[92mAfter~~~~~~~~~~~~~~~~\033[0m")
            print(temp)
            print()
    # break
```

## Preprocessing

```python
os.makedirs('Q1_2_output', exist_ok=True)
base="Q1_1_output"
sample=os.listdir(base)[:5]
for filename in os.listdir(base):
    with open(base+"/"+filename, 'r+') as file:
        content = file.read()
        content1 = content.lower()
        tokens = word_tokenize(content1)
        # print(len(tokens))
        stop_words = set(stopwords.words("english"))
        tokens1 = [token for token in tokens if token not in stop_words]
        # print(len(tokens))
        punctuation = set(string.punctuation)
        tokens2 = [token for token in tokens1 if token not in punctuation] 
        # print(len(tokens))
        tokens3 = [token for token in tokens2 if token.strip()]
        # print(len(tokens))
        with open("Q1_2_output"+"/"+filename, 'w') as file2:
            file2.write(str(tokens3))
        if filename in sample:
            print("\033[91mBefore~~~~~~~~~~~~~~~\033[0m")
            print(content)
            print("\033[92mAfter lowercase~~~~~~~~~~~~~~~\033[0m")
            print(content1)
            print("\033[92mAfter tokenization~~~~~~~~~~~~~~~\033[0m\n")
            print(tokens)
            print("\n\033[92mAfter removing stopwords~~~~~~~~~~~~~~~\033[0m\n")
            print(tokens1)
            print("\n\033[92mAfter removing punctuations~~~~~~~~~~~~~~~\033[0m\n")
            print(tokens2) 
            print("\n\033[92mAfter removing blankspace tokens~~~~~~~~~~~~~~~\033[0m\n")
            print(tokens3)
            print()
        # break
```
